##### **---> Bonus (Función de color normalizada para la paleta de color):**

Con el fin de tener una escala de color que sea acorde a las magnitudes que se presentan en cada una de las gráficas en las secciones asociadas a los EDA, incluiremos como bonus esta función que nos permite tener una paleta normalizada acore los propiso valores de la distribución que toma la variable en estudio. Los colores cálidos indican valores por encima de los valores medios, y los colores fríos valores que estén por debajo. Cuanto más cálidos o más fríos, mayor es la magnitud de la variable. Esto no permite tener una lectura más intuitiva de la distribución.

````python
def colors_from_values(values, palette_name):
    # normalize the values to range [0, 1]
    normalized = (values - min(values)) / (max(values) - min(values))
    # convert to indices
    indices = np.round(normalized * (len(values) - 1)).astype(np.int32)
    # use the indices to get the colors
    palette = sns.color_palette(palette_name, len(values))
    return np.array(palette).take(indices, axis=0)
````

A continuación presentamos la función que permite generar la tendencia del valor medio de la ventana móvil, y las bandas superiores en inferiores que determinarán los valores que se entenderán como outliers cuando se excedan:

````python
def plotMovingAverage(series, window, plot_intervals=False, scale=1.96, 
                      plot_anomalies=False):

    """
        series - dataframe with timeseries
        window - rolling window size 
        plot_intervals - show confidence intervals
        plot_anomalies - show anomalies 

    """
    global rolling_mean
    rolling_mean = series.rolling(window=window).mean()

    plt.figure(figsize=(16,10))
    plt.plot(series[window:], label="Actual values")
    plt.title("Moving average\n window size = {}".format(window))
    plt.plot(rolling_mean, "yellow", label="Rolling mean trend")

    # Plot confidence intervals for smoothed values
    if plot_intervals:
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        deviation = np.std(series[window:] - rolling_mean[window:])
        lower_bond = rolling_mean - (mae + scale * 1.5*deviation)
        upper_bond = rolling_mean + (mae + scale * 1.5* deviation)
        plt.plot(upper_bond, "r--", label="Upper Bond / Lower Bond")
        plt.plot(lower_bond, "r--")
        
        # Having the intervals, find abnormal values
        if plot_anomalies:
            global anomalies
            anomalies = pd.DataFrame(index=series.index, columns=series.columns)
            anomalies[series<lower_bond] = series[series<lower_bond]
            anomalies[series>upper_bond] = series[series>upper_bond]
            plt.plot(anomalies, "ro", markersize=10)
        
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.grid(True)
````

Graficando la salida de la función evidenciamos gráficamente los valores considerados outliers:
